```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
```

```python
# access GISS data copied from:
# https://portal.nccs.nasa.gov/datashare/giss_cmip6/DAMIP/NASA-GISS/GISS-E2-2-G/
# e.g., hist-sol/r1i1p3f1/Amon/tas/gn/v20210911/tas_Amon_GISS-E2-2-G_hist-sol_r1i1p3f1_gn_197601-200012.nc

ds = xr.open_dataset('../data/toz_AERmon_GISS-E2-2-G_hist-sol_r1i1p3f1_gn_197601-200012.nc', decode_times=True)
ds0 = xr.open_dataset('../data/toz_AERmon_GISS-E2-2-G_historical_r1i1p3f1_gn_197601-200012.nc', decode_times=True)
```

```python
toz=ds['toz'].mean(dim='lon').sel(lat=0, method='nearest')
toz0=ds0['toz'].mean(dim='lon').sel(lat=0, method='nearest')
```
